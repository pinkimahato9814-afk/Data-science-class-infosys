# Hypothesis Testing

```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('Dataset/pizza_sales.csv')
```

## One Sample-T Test

### Is the average total_price of all pizza sales different from $15?

```python
'''
alpha value -> Risk Limit -> 5% (0.05)
actual population (column)
sample mean
calculation
Print values (Ho, Ha, actual mean, sample mean)
compare p_value and alpha value
'''
```

```python
alpha = 0.05
price = df['total_price']
sample_mean = 2
```

```python
# (actual_mean - sample_mean) / ((sd)/root(total_no_of_record))
# If statistics value is close to 0 then p_value will be more than 0.5.
# If statistics value is far more than 0 then p_value will be 0.

# import numpy as np
# actual_mean = df['total_price'].mean()
# sd = df['total_price'].std()
# root = np.sqrt(48619)

# top_value = (actual_mean - sample_mean)
# bottom_value = (sd/root)
# top_value / bottom_value
```

```python
# stats.ttest_1samp(population, popmean=sample_mean)
t_statistics, p_value = stats.ttest_1samp(price, popmean=sample_mean)
```

```python
ho = 'The average total price of all pizza sales are equal to $15.'
ha = 'The average total price of all pizza sales are not equal to $15.'
print(f'Ho: {ho}')
print(f'Ha: {ha}')
print(f'Actual Average Price: ${price.mean():,.2f}')
print(f'Sample Average Price: ${sample_mean}')
print(f'Statistics: {t_statistics}')
print(f'Probability Value: {p_value}')
```

```python
if p_value > alpha:
    print(f'Decision: Failed to Reject null hypothesis.\nConclusion: Needs to accept Null Hypothesis.\n{ho}')
else:
    print(f'Decision: Failed to Accept null hypothesis.\nConclusion: Needs to accept Alternate Hypothesis.\n{ha}')
```

## Two Sample-T Test

### Is the average unit price of Classic Pizzas and Chicken Pizzas are Equal?

```python
alpha=0.05
classic_price = df[df['pizza_category'] == 'Classic']['unit_price']
chicken_price = df[df['pizza_category'] == 'Chicken']['unit_price']
```

```python
# stats.ttest_ind(first_value, second_value, equal_var=False)
t_statistics, p_value = stats.ttest_ind(chicken_price, classic_price, equal_var=False)
```

```python
ho = 'The average unit price of Classic Pizzas and Chicken Pizzas are Equal.'
ha = 'The average unit price of Classic Pizzas and Chicken Pizzas are not Equal.'
print(f'Ho: {ho}')
print(f'Ha: {ha}')
print(f'Average Classic Pizza Price: ${classic_price.mean():,.2f}')
print(f'Average Chicken Pizza Price: ${chicken_price.mean():,.2f}')
print(f'T-Statistics: {t_statistics}')
print(f'Probability Value: {p_value}')
```

```python
if p_value > alpha:
    print(f'Decision: Failed to Reject null hypothesis.\nConclusion: Needs to accept Null Hypothesis.\n{ho}')
else:
    print(f'Decision: Failed to Accept null hypothesis.\nConclusion: Needs to accept Alternate Hypothesis.\n{ha}')
```

## ANOVA (Analysis of Variance) Test

### The average total_price generated by all four pizza categories (Classic, Chicken, Supreme, and Veggie) differ significantly from one another?

```python
alpha=0.05
grouped_price = []
```

```python
categories = ['Classic', 'Chicken', 'Supreme', 'Veggie']
for category in categories:
    price = df[df['pizza_category'] == category]['total_price']
    grouped_price.append(price)
```

```python
t_statistics, p_value = stats.f_oneway(*grouped_price)
```

```python
ho = 'The average total price generated by all four pizza category are equal.'
ha = 'The average total price generated by all four pizza category are not equal.'
print(f'Ho: {ho}')
print(f'Ha: {ha}')
mean_price = df.groupby('pizza_category')['total_price'].mean()
format_price = mean_price.map("${:,.2f}".format)
print('-'*36)
print(format_price.to_markdown(numalign='center', stralign='center'))
print('-'*36)
```

```python
if p_value > alpha:
    print(f'Decision: Failed to Reject null hypothesis.\nConclusion: Needs to accept Null Hypothesis.\n{ho}')
else:
    print(f'Decision: Failed to Accept null hypothesis.\nConclusion: Needs to accept Alternate Hypothesis.\n{ha}')
```

# two sample t test
### Is the average  unit price of classic pizza  and cheken pizza is equal ?

```python
alpha = 0.05

classic_price  = df[df['pizza_category']=='Classic']['unit_price']
chicken_price  = df[df['pizza_category']=='Chicken']['unit_price']
```

```python
stats.ttest_ind( classic_price,chicken_price,equal_var = False)
```

```python
# stats.ttest_ind( first_value,second_value,equal_var)
t_statisticsz,p_value = stats.ttest_ind( classic_price,chicken_price,equal_var = False)
```

```python
ho = 'the average price of classic_pizza and  checken pizza is equal'
ha = 'the average price of classic_pizza and  checken pizza is not equal'

print(f'HO: {ho}')
print(f'Ha: {ha}')
print(f'average classic pizza price : ${classic_price.mean():,.2f}')
print(f'average chicken pizza price : ${chicken_price.mean():,.2f}')
print(f'probability value :{p_value}')
print(f't_statistics value :{t_statisticsz}')



```

```python
if p_value > alpha:
    print(f'Decision: failed to reject null hypothesis \n conclusion : need to accept null hypothesis \n .{ho}')
else:
    print(f'Decision: failed to accept null hypothesis \n conclusion : need to accept alternative hypothesis \n .{ha}')
    
```

# Anova testing
 

```python
df['pizza_category'].unique()
```

```python
alpha  =  0.05
grouped_price  = []
categories = ['Classic','Chicken','Supreme','Veggie']
for  category in categories:
    price  = df[df['pizza_category']==category]['total_price']
    grouped_price.append(price)
```

```python
grouped_price
```

```python
stats.f_oneway(*grouped_price)
```

```python

F_statistic ,p_value =  stats.f_oneway(*grouped_price)

print("F-statistic:", F_statistic)
print("p_value:", p_value)
```

```python
ho = 'The average total price generated by all four pizza category are equal.'
ha = 'The average total price generated by all four pizza category are not equal.'
print(f'Ho: {ho}')
print(f'Ha: {ha}')
mean_price = df.groupby('pizza_category')['total_price'].mean()
format_price = mean_price.map("${:,.2f}".format)
print('-'*36)
print(format_price.to_markdown(numalign='center', stralign='center'))
print('-'*36)
```

```python
alpha = 0.05

if p_value > alpha:
    print(f'Decision: failed to reject null hypothesis\n conclusiosn : need to accept null  hypothesis\n . {ho}')
else:
    print(f'Decision: failed to accept null hypothesis\n conclusiosn : need to accept alternative hypothesis\n . {ha}')
```

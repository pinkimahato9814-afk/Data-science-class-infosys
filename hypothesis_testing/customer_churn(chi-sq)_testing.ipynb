```python
from scipy.stats import chi2_contingency
```

# chi-square test

```python
alpha  = 0.05
grouped_value = []
```

```python
features = ["gender", "Contract", "PaymentMethod"]
```


The feature is dependent of churn.
- Gender affects Churn.
- Contract affects Churn.
- PaymentMethod affects Churn.



The feature is independent of churn.
- Gender does not affect Churn.
- Contract does not affect Churn.
- PaymentMethod does not affect Churn.

```python
ho = 'The feature is dependent of churn.'
ha = 'The feature is independent of churn.'
print(f"Null Hypothesis: {ho}")
print(f"Alternate Hypothesis: {ha}")

```

```python
results = []

for feature in features:
    table = pd.crosstab(df[feature], df[target])
    
    chi2, p_value, dof, expected = chi2_contingency(table)
```

```python
chi2_contingency(table)
```

```python

if p_value > alpha:
    print(f'Decision: failed to reject null hypothesis\n conclusiosn : need to accept null  hypothesis\n . {ho}')
else:
    print(f'Decision: failed to accept null hypothesis\n conclusiosn : need to accept alternative hypothesis\n . {ha}')
```

```python
results = []

for feature in features:
    table = pd.crosstab(df[feature], df[target])
    
    chi2, p_value, dof, expected = chi2_contingency(table)
    
    if p_value < alpha:
        decision = "Dependent on Churn"
        important = "Yes"
    else:
        decision = "Independent of Churn"
        important = "No"
    
    results.append({
        "Feature": feature,
        "Chi-Square Value": chi2,
        "p-value": p_value,
        "Degree of Freedom": dof,
        "Decision": decision,
        "Important Feature": important
    })

results_df = pd.DataFrame(results)
results_df
```

```python
import matplotlib.pyplot as plt
import numpy as np  
import seaborn as sns
```

```python
fruits = ['apple','banana','orange','guva','graps']
sales = [500,900,300,600,500]

```

```python
plt.figure(figsize=(6, 4))

plt.bar(fruits, sales, 
        color=['#acd32c', '#e3e3e3', '#2596be', '#6BF2D9', 'orange'], 
        width=0.7)

plt.xlabel('Fruits')
plt.ylabel('Price of Fruits')
plt.title('Total Sales of Fruits')

for idx, val in enumerate(sales):
    plt.text(fruits[idx], val, str(val) ,ha ='center',va ='center',color = 'green')
plt.yticks(rotation=45)
plt.show()
```

```python
department_details = df1.groupby('Department').agg(
    total_salary = ('Salary','sum'),
     total_absence = ('Absences','sum'),
      total_empsatis = ('EmpSatisfaction','mean'),
       total_empid = ('EmpID','count'),
).reset_index()
department_details
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
plt.subplots_adjust(hspace=0.5)
 
# Total Salary
ax1 = axes[0, 0]
ax1.bar(department_details['Department'], department_details['total_salary'], color='red', edgecolor='black')
ax1.set_xlabel('Departments')
ax1.set_ylabel('Total Salary')
ax1.set_title('Total Salary by Departments')
ax1.tick_params(axis='x',rotation=15)
 
for idx, val in enumerate(department_details['total_salary']):
    ax1.text(department_details['Department'][idx], val, str(val), ha='center', va='bottom')
 
# Total Absence Rate
ax2 = axes[0, 1]
ax2.bar(department_details['Department'], department_details['total_absence'], color='blue', edgecolor='black')
ax2.set_xlabel('Departments')
ax2.set_ylabel('Total Absences')
ax2.set_title('Total Absence Rate by Departments')
ax2.tick_params(axis='x', rotation=15)
 
for idx, val in enumerate(department_details['total_absence']):
    ax2.text(department_details['Department'][idx], val, str(val), ha='center', va='bottom')
 
# Average Employee Satisfaction
ax3 = axes[1, 0]
ax3.bar(department_details['Department'], department_details['total_empid'], color='green', edgecolor='black')
ax3.set_xlabel('Departments')
ax3.set_ylabel('Average Employee Satisfaction')
ax3.set_title('Average Employee Satisfaction by Departments')
ax3.tick_params(axis='x', rotation=15)
 
for idx, val in enumerate(department_details['total_empid']):
    fmt_val = round(val, 2)
    ax3.text(department_details['Department'][idx], fmt_val, str(fmt_val), ha='center', va='bottom')
 
# Total Employees
ax4 = axes[1, 1]
ax4.bar(department_details['Department'], department_details['total_empid'], color='purple', edgecolor='black')
ax4.set_xlabel('Departments')
ax4.set_ylabel('Total Employees')
ax4.set_title('Total Employees by Departments')
ax4.tick_params(axis='x', rotation=15)
 
for idx, val in enumerate(department_details['total_empid']):
    ax4.text(department_details['Department'][idx], val, str(val), ha='center', va='bottom')
 
plt.savefig('subplot.png')
```

```python
# Grouping
department_details = df1.groupby('Department').agg(
    total_salary=('Salary', 'sum'),
    total_absence=('Absences', 'sum'),
    total_empsatis=('EmpSatisfaction', 'mean'),
    total_empid=('EmpID', 'count'),
).reset_index()

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 18))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

# Plot 1: Total Salary
axes[0, 0].bar(
    department_details['Department'],
    department_details['total_salary'],
    color='red',
    edgecolor='black'
)
axes[0, 0].set_title('Total Salary by Department')
axes[0, 0].set_xlabel('Department')
axes[0, 0].set_ylabel('Total Salary')
axes[0, 0].tick_params(axis='x', rotation=45)

# Plot 2: Total Absence
axes[0, 1].bar(
    department_details['Department'],
    department_details['total_absence'],
    color='blue',
    edgecolor='black'
)
axes[0, 1].set_title('Total Absence by Department')
axes[0, 1].set_xlabel('Department')
axes[0, 1].set_ylabel('Total Absence')
axes[0, 1].tick_params(axis='x', rotation=45)

# Plot 3: Employee Satisfaction
axes[1, 0].bar(
    department_details['Department'],
    department_details['total_empsatis'],
    color='green',
    edgecolor='black'
)
axes[1, 0].set_title('Avg Employee Satisfaction')
axes[1, 0].set_xlabel('Department')
axes[1, 0].set_ylabel('Satisfaction')
axes[1, 0].tick_params(axis='x', rotation=45)

# Plot 4: Employee Count
axes[1, 1].bar(
    department_details['Department'],
    department_details['total_empid'],
    color='purple',
    edgecolor='black'
)
axes[1, 1].set_title('Employee Count by Department')
axes[1, 1].set_xlabel('Department')
axes[1, 1].set_ylabel('Number of Employees')
axes[1, 1].tick_params(axis='x', rotation=45)




plt.show()
```

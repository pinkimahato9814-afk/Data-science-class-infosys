```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
# Input (study hours)
study_hours = [1, 2, 3, 4, 5]
X = np.array(study_hours).reshape(-1,1)

# Output (0 = Fail, 1 = Pass)
y = np.array([0, 0, 0, 1, 1])
```

```python
# Create model
model = LogisticRegression()
```

```python
# Train model
model.fit(X, y)
# Take user input
hours = float(input("Enter study hours: "))

# Predict
prediction = model.predict([[hours]])
```

```python
print(prediction)
print("Result:", "Pass" if prediction[0] == 1 else "Fail")
```

```python
# Calculate log loss
log_loss = -np.mean(
    y * np.log(prediction) + (1 - y) * np.log(1 - prediction)
)

print("Log Loss:", log_loss)
```

# Error calculation in logistic regression


```python
import numpy as np
from sklearn.linear_model import LogisticRegression

# Input (study hours)
study_hours = [1, 2, 3, 4, 5]
X = np.array(study_hours).reshape(-1, 1)

# Output (0 = Fail, 1 = Pass)
y = np.array([0, 0, 0, 1, 1])

# Create and train model
model = LogisticRegression()
model.fit(X, y)

# Take user input
hours = float(input("Enter study hours: "))

# ❗ Use predict_proba instead of predict
prob = model.predict_proba([[hours]])[0][1]

# Convert to class
prediction = 1 if prob >= 0.5 else 0

print("Probability:", prob)
print("Result:", "Pass" if prediction == 1 else "Fail")

# ✅ Calculate log loss (for this prediction)
epsilon = 1e-9  # avoid log(0)
prob = np.clip(prob, epsilon, 1 - epsilon)

actual = float(input("Enter actual result (0 or 1): "))

log_loss = -(actual * np.log(prob) + (1 - actual) * np.log(1 - prob))

print("Log Loss:", log_loss)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
study_hours = [1,2,4,5,6,8,10]
exam_score = [20,30,40,50,70,80,95]
x= [[i] for i in  study_hours]
y = exam_score

model= LinearRegression()
model.fit(x,y)
std_hr = int(input("enter your study hours: "))
predict = model.predict([[std_hr]])
print(f"score: {round(predict[0],0)}%")
```

#### Intercept = predicted score when study hours = 0
#### Slope = how much the score increases for each extra hour of study
##### “Even without studying, the student might score ~25 (basic knowledge, guessing, prior learning).”


```python
#Slope = learning speed 📈
#Intercept = starting point
# check model
print("Slope (m):", model.coef_)
print("Intercept (b):", model.intercept_)
```

### BY using Rshape --for(1DA -->2DA)

```python
import numpy as np
from sklearn.linear_model import LinearRegression

study_hours = [1,2,4,5,6,8,10,12]
exam_score = [20,30,40,50,70,80,95,100]

# Convert to numpy arrays and reshape
x = np.array(study_hours).reshape(-1, 1)
y = np.array(exam_score)
```

```python
model = LinearRegression()
model.fit(x, y)

std_hr = int(input("enter your study hours: "))
predict = model.predict([[std_hr]])

print(f"score: {round(predict[0], 0)}%")
```

```python
print(f"intercept(b) :{model.intercept_}") 
print(f"slop(m) : {model.coef_} " )
```

# Error calculation in our prediction model



```python
import numpy as np

# actual and predicted values
y_true = np.array([y])
y_pred = np.array([predict])

# calculate MSE
mse = np.mean((y_true - y_pred) ** 2)

print("MSE:", mse)
```

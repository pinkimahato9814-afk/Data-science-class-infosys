```python
department_details = df.groupby('Department').agg(
    total_salary = ('Salary','sum'),
    total_employee = ('DeptID','sum')

).reset_index()


department_details 
```

```python
# Print report of an analysis where total salary, total employees, average employee satisfaction,
# total absence days of emplyoees by each department.
# agg() -> aggregate function


dep_details = df.groupby('Department').agg(
    Total_salary = ('Salary','sum'),
    Total_employee = ('EmpID','sum'),
    Avg_emp_satisfaction =("EmpSatisfaction","sum"),
    Total_absence = ("Absences","sum")

).reset_index()
dep_details
```

```python
dep_details = df.groupby('Department').agg(
    Total_salary = ('Salary','sum'),
    Total_employee = ('EmpID','sum'),
    Avg_emp_satisfaction =("EmpSatisfaction","sum"),
    Total_absence = ("Absences","sum")

).sort_values(ascending= True, by = 'Total_salary').reset_index()

dep_details['Tota_salary'] = dep_details['Total_salary'].map("rs.{:,.2f}".format)
dep_details['Avg_emp_satisfaction'] = dep_details['Avg_emp_satisfaction'].map("rs.{:,.2f}".format)
dep_details

```

```python
# Find total employees data by each department and its performance score.
total_employee = df.groupby(['Department','PerformanceScore'])['EmpID'].count().reset_index()
total_employee

```

```python
# find total Employees data by each department and performance score 
dep_perf = df.groupby(['Department','PerformanceScore'])['EmpID'].count().reset_index()
dep_filter = (dep_perf['PerformanceScore']=='Fully Meets')|(dep_perf['PerformanceScore']=='Exceeds')
dep_perf[dep_filter]
```

```python
## find total Employees data by each department and performance score 
# same problem by using isin function
dep_filter = dep_perf['PerformanceScore'].isin(['Fully Meets','Exceeds'])
dep_perf[dep_filter]
```

```python

df['DateofHire'] = pd.to_datetime(df['DateofHire'])
```

```python
df['Hire_Year'] = df['DateofHire'].dt.year
df['Hire_Month_name'] = df['DateofHire'].dt.month_name()
df['Hire_Day_name'] = df['DateofHire'].dt.day_name()
```

```python
df['Hire_Day_name'].unique()
```

```python
weekdays = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
# Categorical -> pandas
df['Hire_Day_name'] = pd.Categorical(df['Hire_Day_name'], categories=weekdays)
df['Hire_Day_name']
```

```python
# Find total employees hired in each day.
hire_day = df.groupby('Hire_Day_name', observed=False)['EmpID'].count().reset_index()
hire_day
```
